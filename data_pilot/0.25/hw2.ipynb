```python
import os
```

```python
for a in ["quick", "bubble"]:
    print("Using:", a)
    for i in [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]:
        print("n:", i)
        output = a+"_"+str(i)+".txt"
        for j in range(20):
            print(j,"/",20)
            os.system("python sort.py -n "+str(i)+" -a "+a+" >> "+output)
```
